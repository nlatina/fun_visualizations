```python
def wordle(guesses: list[str], green: dict, yellow: dict) -> list[str]:
    import re
    
    chars = set('abcdefghijklmnopqrstuvwxyz')

    for g in guesses:
        chars -= set(g)
    

    
    remaining_letters = ''
    for c in chars:
        remaining_letters += c
        
    remaining_letters += ''.join([str(value) for value in green.values()])
    
    if yellow:
        yellow_letters = set()
        for value_list in yellow.values():
            for v in value_list:
                yellow_letters.add(v)
        remaining_letters += ''.join([str(c) for c in yellow_letters])####

    pattern = fr'[{remaining_letters}]{{5}}'
    finder = re.compile(pattern)
    
    if green:
        filtr = [w for w in words if 
                    all(w[key] == value for key, value in green.items())
                   and finder.match(w)] 
    else:
        filtr = [w for w in words if finder.match(w)]
        
    if yellow:
        for item in yellow.items():
            filtr = [w for w in filtr if all(w[item[0]] != value for value in item[1])]

        filtr = [w for w in filtr if all(c in w for c in yellow_letters)]
        
    
    
    return filtr
    
guesses = ['ocean']
green = {0:'o', 4:'n'}
yellow = {2:'e'}

guesses = wordle(guesses, green = green, yellow = yellow)

import random
print(guesses)
print(random.choice(guesses))
```
